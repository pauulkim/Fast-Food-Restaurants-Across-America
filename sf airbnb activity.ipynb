```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# EDA

## Neighborhoods

#### Price Distribution

```python
plt.figure(figsize=(20,20))

# boxplot on price and neighborhood with averages
sns.boxplot(x="price", y="neighbourhood_cleansed", data=listings, showfliers=False, showmeans=True, 
            meanprops={'marker':"d","markerfacecolor":"red","markeredgecolor":"white","markersize":10})

# plot the average overall price 
plt.axvline(x=listings["price"].mean(), color="crimson")

plt.xlabel("Price")
plt.ylabel("Neighborhood")
plt.show()
```

#### Overall Review Rating Distribution

```python
plt.figure(figsize=(20,20))

# boxplot on review scores and neighborhood with averages
sns.boxplot(x="review_scores_rating", y="neighbourhood_cleansed", data=listings, showfliers=False, showmeans=True, 
            meanprops={'marker':"d","markerfacecolor":"red","markeredgecolor":"white","markersize":10})

# plot the average overall review score 
plt.axvline(x=listings["review_scores_rating"].mean(), color="crimson")

plt.xlabel("Review Ratings")
plt.ylabel("Neighborhood")
plt.show()
```
